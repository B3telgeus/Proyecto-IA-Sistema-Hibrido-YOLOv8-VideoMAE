```python
import cv2
from ultralytics import YOLO
import gradio as gr

# Carga del modelo preentrenado (Baseline)
# Usamos 'yolov8n.pt' por su alta velocidad para operación en tiempo real
model = YOLO('yolov8n.pt')

def predict_and_plot(img):
    """
    Función que recibe una imagen, realiza la detección y
    devuelve la imagen con los Bounding Boxes dibujados.
    """
    # Realizar la detección
    results = model.predict(source=img, conf=0.25)

    # El método .plot() dibuja automáticamente las detecciones en la imagen
    res_plotted = results[0].plot()

    # Convertir de BGR (OpenCV) a RGB (Gradio)
    return cv2.cvtColor(res_plotted, cv2.COLOR_BGR2RGB)
```

```python
# Configuración de la interfaz de Gradio
demo = gr.Interface(
    fn=predict_and_plot,
    inputs=gr.Image(type="numpy", label="Subir imagen de estación/bus"),
    outputs=gr.Image(type="numpy", label="Detección de Seguridad"),
    title="Prototipo IA - Seguridad SITP/TransMilenio",
    description="Demo temprana para detección de personas y zonas críticas utilizando YOLOv8."
)

# Launch con share=True genera un link público válido por 72 horas
demo.launch(share=True)
```
